```python
import pickle
import torch
import os
import numpy as np
import pandas as pd
```

```python
seed = 0
vocab_size = 50

result_table = {'composition_fn': [], 'message_len': [], 'values': [], 'lr': [], 'weight-decay': [] ,
                'train_min': [], 'val_min': []}

for composition_function in ['linear', 'mlp']:
    for message_length in [4,6]:
        for values in [16, 50]:
            for learning_rate in [0.1, 0.01, 0.001]: 
                for weight_decay in [1e-1, 1e-3]:
                    
                    result_table['composition_fn'].append(composition_function)
                    result_table['message_len'].append(message_length)
                    result_table['values'].append(values)
                    result_table['lr'].append(learning_rate)
                    result_table['weight-decay'].append(weight_decay)
                
                    path = ('results/grid_search/atts2_vals' + str(values) + '_vs50_len' + str(message_length) 
                            + '_compfn-' + composition_function + '_lr' + str(learning_rate) + 
                            '_weight-decay' + str(weight_decay))

                    tre = pickle.load(open(path + '_TRE.pkl', 'rb'))['seed0']
                    index = np.argmin(tre['validation_mean'])
                    result_table['train_min'].append(tre['training_mean'][index])
                    result_table['val_min'].append(tre['validation_mean'][index])
                    
df = pd.DataFrame(data=result_table)
```
